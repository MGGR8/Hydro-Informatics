```python
import matplotlib.pyplot as plt
import numpy as np

def plot_values(names, a_values, b_values, c_values):
    # Number of individuals
    num_bars = len(names)

    # Position for each bar on the x-axis
    x_pos = range(num_bars)

    # Create the plot
    fig, ax = plt.subplots()

    # Create bars
    bars_a = ax.bar(x_pos, a_values, color='blue', label='a')
    bars_b = ax.bar(x_pos, b_values, color='orange', label='b', bottom=a_values)
    bars_c = ax.bar(x_pos, c_values, color='green', label='c', bottom=[a + b for a, b in zip(a_values, b_values)])

    # Set the y-axis label
    ax.set_ylabel('Values')

    # Set the title
    ax.set_title('Values by Name')

    # Set the position of the x ticks
    ax.set_xticks(x_pos)

    # Set the labels for the x ticks
    ax.set_xticklabels(names)

    # Add a legend
    ax.legend()

    # Show the plot
    plt.show()
    
def plot_values2(names, a_values, b_values, c_values):
    # Create the plot
    fig, ax = plt.subplots()

    # Plot lines for each value
    ax.plot(names, a_values, marker='o', label='a')
    ax.plot(names, b_values, marker='o', label='b')
    ax.plot(names, c_values, marker='o', label='c')

    # Set the x-axis label
    ax.set_xlabel('Names')

    # Set the y-axis label
    ax.set_ylabel('Values')

    # Set the title
    ax.set_title('Values by Name')

    # Rotate x-axis labels for better readability if there are many names
    plt.xticks(rotation=45, ha='right')

    # Add a legend
    ax.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()
    
def plot_values3(names, a_values, b_values, c_values):
    # Number of individuals
    num_bars = len(names)

    # Width of each bar
    bar_width = 0.25

    # Position for each group on the x-axis
    index = np.arange(num_bars)

    # Create the plot
    fig, ax = plt.subplots()

    # Create bars for each value
    ax.bar(index - bar_width, a_values, bar_width, color='blue', label='a')
    ax.bar(index, b_values, bar_width, color='orange', label='b')
    ax.bar(index + bar_width, c_values, bar_width, color='green', label='c')

    # Set the x-axis label
    ax.set_xlabel('Names')

    # Set the y-axis label
    ax.set_ylabel('Values')

    # Set the title
    ax.set_title('Values by Name')

    # Set the position of the x ticks
    ax.set_xticks(index)

    # Set the labels for the x ticks
    ax.set_xticklabels(names)

    # Add a legend
    ax.legend()
    
    plt.xticks(rotation=45, ha='right')

    # Show the plot
    plt.tight_layout()
    plt.show()

def read_data_from_file(filename):
    names = []
    a_values = []
    b_values = []
    c_values = []

    with open(filename, 'r') as file:
        next(file)  # Skip header line
        for line in file:
            data = line.strip().split(',')
            names.append(data[0])
            a_values.append(int(data[1]))
            b_values.append(int(data[2]))
            c_values.append(int(data[3]))

    return names, a_values, b_values, c_values

# Example usage
filename = 'data.txt'  # Change this to your filename
names, a_values, b_values, c_values = read_data_from_file(filename)
plot_values(names, a_values, b_values, c_values)
plot_values2(names, a_values, b_values, c_values)
plot_values3(names, a_values, b_values, c_values)
```
